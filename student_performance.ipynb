# Veri Seti Tanımı ve Sütun Açıklamaları

Bu veri seti, öğrencilerin akademik performansını etkileyen çeşitli faktörleri içermektedir.

| Sütun Adı | Açıklama | Veri Tipi |
|---|---|---|
| Student_ID | Öğrenciye özel benzersiz numara | Sayısal (ID) |
| Weekly_Study_Hours | Haftalık ders çalışma süresi (saat) | Sayısal |
| Attendance_Rate | Derslere katılım oranı (%) | Sayısal |
| Sleep_Hours | Günlük ortalama uyku süresi | Sayısal |
| Previous_Scores | Önceki sınavların not ortalaması | Sayısal |
| Internet_Usage_Time | Günlük internet kullanım süresi (saat) | Sayısal |
| Parent_Education_Level | Ebeveyn eğitim seviyesi | Kategorik / Sıralı |
| Distance_to_School | Ev ile okul arasındaki mesafe (km) | Sayısal |
| Stress_Level | Öğrencinin stres seviyesi (1-10) | Sayısal |
| Practice_Exams_Passed | Başarıyla geçilen deneme sınavı sayısı | Sayısal |
| Extracurricular_Activities | Sosyal aktivitelere ayrılan süre | Sayısal |
| **Exam_Score** | **Final sınavı notu** | Sayısal |

**Kütüphanelerin Yüklenmesi**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
print("kütüphaneler başarıyla yüklendi.")
```

**Veri Setinin Yüklenmesi**

```python
# Veri setini yükle
df = pd.read_csv('dataset/student_performance.csv')

print("Veri seti başarıyla yüklendi!")
```

**Veriye İlk Bakış**

```python
print("--- İlk 5 Satır ---")
display(df.head())

print("\n--- Son 5 Satır ---")
display(df.tail())

print("\n--- Rastgele 5 Örnek ---")
display(df.sample(5))
```

**Veri Yapısı**

```python
print("--- Veri Bilgisi (Info) ---")
df.info()

print("\n--- İstatistiksel Özet ---")
display(df.describe().T)
```

**Eksik Veri Kontrolü**

```python
# --- Eksik Veri Analizi: Sayısal ve Görsel Karşılaştırma ---

# 1. Adım: Eksik Veri Sayılarını Hesaplama
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

# 2. Adım: Yan Yana Grafikler (Subplots)
if not missing_data.empty:
    # 1 satır, 2 sütunluk bir figür oluşturuyoruz
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))

    # SOL GRAFİK: Isı Haritası (Heatmap)
    # Verideki boşlukların genel dağılımını gösterir
    sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis', ax=axes[0])
    axes[0].set_title('1. Eksik Veri Matrisi (Genel Görünüm)', fontsize=14)

    # SAĞ GRAFİK: Sütun Grafiği (Bar Plot)
    # Hangi sütunda kaç tane eksik veri olduğunu gösterir
    sns.barplot(x=missing_data.index, y=missing_data.values, palette='mako', ax=axes[1])
    axes[1].set_title('2. Sütun Bazlı Eksik Veri Sayıları', fontsize=14)
    axes[1].set_ylabel('Eksik Değer Adedi')
    plt.xticks(rotation=45)

    # Sağdaki grafikte çubukların üzerine sayısal değerleri ekleme
    for i, v in enumerate(missing_data.values):
        axes[1].text(i, v + 5, str(v), ha='center', fontweight='bold', color='black')

    plt.tight_layout() # Grafiklerin birbirine girmesini engeller
    plt.show()
else:
    print("Tebrikler! Veri setinde hiç eksik değer bulunamadı.")
```

**Aykırı Veri Kontrolü**

```python
# Sayısal sütunları seçelim (ID hariç)
num_cols = ['Weekly_Study_Hours', 'Attendance_Rate', 'Sleep_Hours', 
            'Previous_Scores', 'Internet_Usage_Time', 'Distance_to_School', 
            'Stress_Level', 'Practice_Exams_Passed', 'Exam_Score']

# Grafiklerin yerleşimi
plt.figure(figsize=(20, 15))

for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(data=df, y=col, color='salmon')
    plt.title(f'{col} Aykırı Değer Kontrolü', fontsize=12)

plt.tight_layout()
plt.show()
```

**Mantıksal Hata Kontrolü**

```python
print("--- Mantıksal Hata Kontrolü ---")

# 1. Uyku saati hatası (24 saatten fazla olamaz)
outlier_sleep = df[df['Sleep_Hours'] > 24]
print(f"Hatalı Uyku Saati Kaydı ( >24 ): {len(outlier_sleep)}")

# 2. Mesafe hatası (Negatif olamaz)
outlier_distance = df[df['Distance_to_School'] < 0]
print(f"Hatalı Mesafe Kaydı ( <0 ): {len(outlier_distance)}")

# 3. IQR Yöntemi ile İstatistiksel Aykırı Değer Tespiti (Örnek: Exam_Score için)
Q1 = df['Exam_Score'].quantile(0.25)
Q3 = df['Exam_Score'].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

outliers_iqr = df[(df['Exam_Score'] < alt_sinir) | (df['Exam_Score'] > ust_sinir)]
print(f"Exam_Score için IQR Yöntemiyle Belirlenen Aykırı Değer Sayısı: {len(outliers_iqr)}")

# Hatalı örnekleri görelim
if len(outlier_sleep) > 0:
    print("\nUyku Saati Hatalı Olan Bazı Kayıtlar:")
    display(outlier_sleep.head())
```

**Sayısal Değişkenlerin Analizi**

```python
# Sayısal değişkenlerin listesi
num_features = ['Weekly_Study_Hours', 'Attendance_Rate', 'Sleep_Hours', 
                'Previous_Scores', 'Internet_Usage_Time', 'Distance_to_School']

plt.figure(figsize=(20, 12))
for i, col in enumerate(num_features):
    plt.subplot(2, 3, i + 1)
    # KDE (Kernel Density Estimate) verinin yoğunluğunu daha yumuşak gösterir
    sns.kdeplot(df[col], fill=True, color='darkorange')
    plt.title(f'{col} Dağılımı', fontsize=13)
    plt.grid(True, alpha=0.7)

plt.tight_layout()
plt.show()
```

Bu grafikte tam olarak yüzdesel hesap yapmak yerine öğrencilerin özelliklerinin yoğunluğu nerede gibi bir bakış açısıyla bakıyoruz. 

Mesela internet kullanımında öğrencilerin çoğu 7-9 saat arasında internet kullandığı için grafik orada yukarı doğru sivrilmiş gözüküyor.

**Kategorik Değişkenlerin Analizi**

```python
# Kategorik ve Kesikli Değişkenler Analizi
cat_features = ['Parent_Education_Level']

plt.figure(figsize=(16, 7))

for i, col in enumerate(cat_features):
    plt.subplot(1, 2, i + 1)
    
    # Veriyi hesapla
    counts = df[col].value_counts().sort_index()
    
    # Grafik çizimi
    ax = sns.barplot(x=counts.index, y=counts.values, palette='viridis')
    
    plt.ylim(0, counts.max() * 1.15)
    plt.title(f'{col} Frekans Analizi', fontsize=14, pad=20)
    plt.ylabel('Öğrenci Sayısı')
    
    for j, val in enumerate(counts.values):
        ax.text(j, val + (counts.max() * 0.02), str(val), 
                ha='center', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
```

Ebeveyn eğitim seviyesinde her durum ortalama sayılarda gözüküyor.

Başarılı geçen deneme sınavı sayılarında 8-11 deneme sınavı sayısında yoğunluk olduğu gözüküyor.

**Hedef Değişken Analizi**

```python
# Hedef Değişkenin Dağılımı ve Merkezi Eğilimleri
plt.figure(figsize=(14, 6))

# Histogram ve Yoğunluk
plt.subplot(1, 2, 1)
sns.histplot(df['Exam_Score'], kde=True, color='royalblue', bins=30)
plt.title('Sınav Notu Dağılımı (Histogram)', fontsize=14)

# Kutu Grafiği (Aykırı değerleri ve çeyreklikleri görmek için)
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Exam_Score'], color='lightsteelblue')
plt.title('Sınav Notu Değişkenliği (Boxplot)', fontsize=14)

plt.tight_layout()
plt.show()

print(f"Sınav Notu Ortalaması: {df['Exam_Score'].mean():.2f}")
print(f"Sınav Notu Standart Sapması: {df['Exam_Score'].std():.2f}")
```

**Korelasyon Analizi**

```python
# --- KORELASYON ANALİZİ ---
plt.figure(figsize=(20, 8))

# 1. Tam Kare Isı Haritası (Full Heatmap)
plt.subplot(1, 2, 1)
# Sadece sayısal sütunları seçip ID'yi çıkarıyoruz
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['Student_ID'], errors='ignore')
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='RdBu', center=0, square=True, cbar_kws={"shrink": .8})
plt.title('Değişkenlerin Tam Korelasyon Matrisi', fontsize=14, pad=20)

# 2. Sınav Notu (Exam_Score) ile Korelasyon Sıralaması
plt.subplot(1, 2, 2)
target_corr = corr_matrix['Exam_Score'].sort_values(ascending=False).drop('Exam_Score')

sns.barplot(x=target_corr.values, y=target_corr.index, palette='coolwarm')
plt.title('Sınav Notunu Etkileyen Faktörler', fontsize=14, pad=20)
plt.xlabel('Korelasyon Katsayısı')

plt.tight_layout()
plt.show()
```

Haftalık çalışma saatinin, sınav notunu, başarılı geçen deneme sınavını ve stres seviyesini ne kadar etkilediği korelasyon analizinde gösterilmiştir.

Derse katılım oranının artmasıyla okul - ev mesafesi ters etkilidir. Çünkü mesafe arttıkça katılım oranı azalabilir. 

**Çok Değişkenli Analiz**

```python
# --- ÇOK DEĞİŞKENLİ ANALİZ ---

# A. JointPlot: Çalışma Saati ve Sınav Notu (Regresyon ve Dağılım Bir Arada)
# Bu grafik yoğunluğu ve trendi aynı anda gösterir
jp = sns.jointplot(data=df, x='Weekly_Study_Hours', y='Exam_Score', kind='reg', 
                  joint_kws={'line_kws':{'color':'red'}}, scatter_kws={'alpha': 0.1})
jp.fig.suptitle('Çalışma Saati - Sınav Notu İlişki Analizi', y=1.03, fontsize=15)
plt.show()

# B. Hue Scatter: Çalışma Saati vs Sınav Notu (Stres Seviyesi Etkisiyle)
# Stres seviyesini görselleştirmeyi kolaylaştırmak için kategorize edelim
df['Stress_Intensity'] = df['Stress_Level'].apply(lambda x: 'Yüksek' if x > 7 else ('Düşük' if x < 4 else 'Orta'))

plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='Weekly_Study_Hours', y='Exam_Score', hue='Stress_Intensity', 
                palette='rocket', alpha=0.5)
plt.title('Çalışma Saati vs Sınav Notu (Stres Gruplarının Etkisi)', fontsize=15, pad=15)
plt.legend(title='Stres Seviyesi', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Çalışma saati arttıkça sınav notunda da gözle görülür bir artış mevcuttur.

Aynı zamanda çalışma saati arttıkça stres seviyesinde de bir artış vardır.

**Yaşam Tarzı ve Alışkanlık Analizi**

```python
# --- YAŞAM TARZI VE ALIŞKANLIK ANALİZİ ---

# Verileri 3 ana gruba (Düşük, Orta, Yüksek) bölüyoruz
df['Internet_Group'] = pd.qcut(df['Internet_Usage_Time'], q=3, labels=['Az', 'Orta', 'Çok'])
df['Sleep_Group'] = pd.qcut(df['Sleep_Hours'], q=3, labels=['Az Uyku', 'Orta Uyku', 'Çok Uyku'])

# Ortalama notları içeren bir Pivot Tablo oluşturma
lifestyle_pivot = df.pivot_table(values='Exam_Score', index='Sleep_Group', columns='Internet_Group', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(lifestyle_pivot, annot=True, cmap='YlGnBu', fmt=".1f", annot_kws={"size": 12})
plt.title('Uyku ve İnternet Kullanımına Göre Ortalama Sınav Notları', fontsize=14, pad=20)
plt.xlabel('İnternet Kullanımı')
plt.ylabel('Uyku Süresi')
plt.show()
```

İnternet kullanımının artması her türlü notlarda başarısızlık olduğunu gösteriyor.

Uyku süresi fazla olanların notları uyku süresinin ortalama ve az olanlara kıyasla az olduğu görülüyor.

**Öğrenci Segmentasyon Analizi**

```python
# --- SEGMENTASYON ANALİZİ ---

# Grupları oluşturma
top_10_percent = df.nlargest(int(len(df)*0.1), 'Exam_Score')
bottom_10_percent = df.nsmallest(int(len(df)*0.1), 'Exam_Score')

top_10_percent['Group'] = 'En Başarılı %10'
bottom_10_percent['Group'] = 'En Başarısız %10'
comparison_df = pd.concat([top_10_percent, bottom_10_percent])

# Kıyaslanacak kritik özellikler
features = ['Weekly_Study_Hours', 'Attendance_Rate', 'Previous_Scores', 'Stress_Level']

plt.figure(figsize=(16, 10))
for i, col in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.barplot(data=comparison_df, x='Group', y=col, palette='muted', ci=None)
    plt.title(f'{col}: Başarılı vs Başarısız', fontsize=13)
    plt.ylabel('Ortalama Değer')

plt.tight_layout()
plt.show()
```

Başarılı öğrencilerde haftalık çalışma saati, başarısızlara göre daha fazladır.

Önceki sınavlarda alınan notların yeni sınava önemli bir katkısı vardır, bu da önceki sınavdaki bilginin ne kadar önemli olduğunu gösteriyor.

Derse katılım oranı çok fazla bir önem göstermese de bir disiplin göstergesidir.

Başarılı öğrencilerin stres seviyelerinin başarısızlara göre fazla olması yoğun çalışma temposu ve akademik kaygının sonucu olabilir.
